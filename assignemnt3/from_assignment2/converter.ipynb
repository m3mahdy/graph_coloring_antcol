```python
import os

input_dir = "./data_original"
output_dir = "./data"

os.makedirs(output_dir, exist_ok=True)

for filename in os.listdir(input_dir):
    if not filename.endswith(".col"):
        continue

    input_path = os.path.join(input_dir, filename)
    output_path = os.path.join(output_dir, filename.replace(".col", ""))

    edges = set()
    num_nodes = 0

    with open(input_path, 'r') as f:
        for line in f:
            if line.startswith('c'):
                continue
            elif line.startswith('p'):
                parts = line.strip().split()
                num_nodes = int(parts[2])
            elif line.startswith('e'):
                _, u, v = line.strip().split()
                u, v = int(u) - 1, int(v) - 1
                # store edge in sorted order to avoid duplicates
                if u != v:
                    edges.add(tuple(sorted((u, v))))

    with open(output_path, 'w') as f:
        f.write(f"{num_nodes} {len(edges)}\n")
        for u, v in sorted(edges):
            f.write(f"{u} {v}\n")

    print(f"{filename}: {len(edges)} unique edges written")

print("✅ All files converted successfully (duplicates removed if any)")
```
